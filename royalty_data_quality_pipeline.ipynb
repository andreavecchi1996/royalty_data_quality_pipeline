## Data cleaning

### df_amazon_jan24_raw

```python

df_amazon_jan24_raw.info()
```

```python
null_counts = df_amazon_jan24_raw.isna().sum()
print('null values count:')
print(null_counts[null_counts > 0])
```

```python
# Remove duplicates.
df = df_amazon_jan24_raw.drop_duplicates()

# Drop columns not required in the final schema.
df = df.drop(columns=["ASIN"])

# Cast to string and strip whitespace.
for col in ["ArtistName", "LabelName", "TrackTitle", "ISRC", "AlbumTitle", "CountryCode", "Service_Tier", "CurrencyCode"]:
    df[col] = df[col].astype("string").str.strip()

# DSP and Source columns creation.
df["DSP"] = "Amazon"
df["Source_File"] = "amazon_jan24_raw.csv"

#-----------------------------------------------------------------------------
# Fill missing LabelName using dim_track as a lookup.
isrc_to_label = (
    df_dim_track[["ISRC", "Label"]]
    .assign(ISRC=lambda x: x["ISRC"].astype("string"))
    .set_index("ISRC")["Label"]
)

# Fill missing LabelName using the lookup.
df["LabelName"] = df["LabelName"].fillna(df["ISRC"].map(isrc_to_label))

#---------
# Build a (ArtistName, TrackTitle) to ISRC lookup from dim_track.
# Lowercased on both sides to ensure the matching with the raw data.
artist_track_to_isrc = (
    df_dim_track[["Artist", "Track", "ISRC"]]
    .assign(
        Artist=lambda x: x["Artist"].astype("string").str.lower(),
        Track=lambda x: x["Track"].astype("string").str.lower(),
        ISRC=lambda x: x["ISRC"].astype("string"),
    )
    .set_index(["Artist", "Track"])["ISRC"]
    .to_dict()
)

# Fill missing ISRCs using the lookup.
df["ISRC"] = df["ISRC"].fillna(
    df.apply(lambda row: artist_track_to_isrc.get(
        (row["ArtistName"].lower(), row["TrackTitle"].lower())
    ), axis=1)
)

#-----------------------------------------------------------------------------

# Fix remaining dtypes.
df["DSP"] = df["DSP"].astype("string")
df["Source_File"] = df["Source_File"].astype("string")
df["UPC"] = df["UPC"].astype("Int64").astype("string") #.astype("Int64") to remove the decimal

# Convert dates to datetime.
df["Period_Start"] = pd.to_datetime(df["Period_Start"], dayfirst=True).dt.normalize()
df["Period_End"] = pd.to_datetime(df["Period_End"], dayfirst=True).dt.normalize()

df_amazon_jan24_clean = df
df_amazon_jan24_clean.info()
```
